```python
feat = df_new.drop(columns=['price'],axis=1)
label = df_new["price"]

cols_to_drop = ["latitude", "longitude", "neighbourhood_Amager Vest", "neighbourhood_Amager st", "neighbourhood_Bispebjerg",
               "neighbourhood_Brnshj-Husum", "neighbourhood_Frederiksberg", "neighbourhood_Indre By",
               "neighbourhood_Nrrebro", "neighbourhood_Valby","neighbourhood_Vanlse", "neighbourhood_Vesterbro-Kongens Enghave",
               "neighbourhood_sterbro", "room_type_Shared room"]
df = df.drop(columns=cols_to_drop,axis=1)

df.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(feat, label, test_size=0.3)
```

```python
from sklearn.preprocessing import MinMaxScaler
sc_x = MinMaxScaler()
X_train = sc_x.fit_transform(X_train)
X_test = sc_x.fit_transform(X_test)

print (X_train.shape)
print (X_test.shape)
```

```python
training_accuracy = dict()
test_accuracy = dict()

max_depth_settings = range(1,30) 

for max_depth in max_depth_settings:
    key = str(max_depth)
    tree = DecisionTreeClassifier(random_state=0, max_depth=max_depth)
    tree.fit(X_train, y_train)
    
    value = tree.score(X_train, y_train)
    training_accuracy[key] = value
    
    value = tree.score(X_test, y_test)
    test_accuracy[key] = value
    
print ("Test  Train  Max_Depth")
for key in sorted(test_accuracy, key=test_accuracy.get, reverse=True):
    print ("{:.4f}, {:.4f}, {}"
          .format(test_accuracy[key], training_accuracy[key], key))
        
```

```python
training_accuracy_ran = dict()
test_accuracy_ran = dict()

max_features_settings = range(1,30) 

for max_features in max_features_settings:
    key = str(max_features)
    ran_forest_ran = RandomForestClassifier(n_estimators=max_features, random_state=0)
    ran_forest_ran.fit(X_train, y_train)
    
    value = ran_forest_ran.score(X_train, y_train)
    training_accuracy_ran[key] = value
    
    value = ran_forest_ran.score(X_test, y_test)
    test_accuracy_ran[key] = value
    
print ("Test  Train  Max_features")
for key in sorted(test_accuracy_ran, key=test_accuracy_ran.get, reverse=True):
    print ("{:.4f}, {:.4f}, {}"
          .format(test_accuracy_ran[key], training_accuracy_ran[key], key))
```

```python
training_accuracy_clf = dict()
test_accuracy_clf = dict()

i_settings = range(1,30) 

for i in i_settings:
    key = str(max_features)
    clf = KNeighborsClassifier(n_neighbors=i)
    clf.fit(X_train, y_train)
    
    value = clf.score(X_train, y_train)
    training_accuracy_clf[key] = value
    
    value = clf.score(X_test, y_test)
    test_accuracy_clf[key] = value
    
print ("Test  Train  Max_features")
for key in sorted(test_accuracy_clf, key=test_accuracy_clf.get, reverse=True):
    print ("{:.4f}, {:.4f}, {}"
          .format(test_accuracy_clf[key], training_accuracy_clf[key], key))
```
